```python
# Dependencies and Setup
import pandas as pd
import os
import csv
from sqlalchemy import create_engine

# File to Load 
file_to_load = os.path.join("Google-Playstore-32K - 2019.csv")

# Read Playstore data and store into Pandas data frame
apps_data_2019 = pd.read_csv(file_to_load)
apps_data_2019.head()
```

```python
#drop the duplicate rows
apps_data_2019.drop_duplicates(subset='app_name', inplace=True)
```

```python
del apps_data_2019['price'] 
del apps_data_2019['minimum_version'] 
del apps_data_2019['latest_version'] 
```

```python
apps_data_2019.head()
```

```python
# File to Load 
file_to_load = os.path.join("app 2020 b (better one).csv")

# Read Playstore data and store into Pandas data frame
apps_data_2020 = pd.read_csv(file_to_load)
apps_data_2020.head()
```

```python
#drop the duplicate rows
apps_data_2020.drop_duplicates(subset='app_name', inplace=True)
```

```python
#remove columns not needed
del apps_data_2020['id']
del apps_data_2020['rate_5_pc']
del apps_data_2020['rate_4_pc']  
del apps_data_2020['rate_3_pc']  
del apps_data_2020['rate_2_pc']  
del apps_data_2020['rate_1_pc']
del apps_data_2020['current_version']
del apps_data_2020['requires_android']  
del apps_data_2020['in_app_products']
del apps_data_2020['offered_by']
del apps_data_2020['cost_label'] 
```

```python
apps_data_2020.head()
```

```python
apps_data_2020 = apps_data_2020.rename(columns={'size':'app_size'})
```

```python
pd.read_sql('''
SELECT  d_19.app_name,
        d_19.category,
        d_19.rating,
        d_19.reviews,
        d_19.app_size,
        d_19.installs,
        d_19.content_rating,
		d_19.last_updated
FROM app_data_19 as d_19
LEFT JOIN app_data_20 as d_20
ON (d_19.app_name = d_20.app_name);''', con=engine)
```
